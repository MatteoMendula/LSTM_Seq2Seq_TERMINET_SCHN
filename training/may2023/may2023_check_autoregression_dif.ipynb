# CHECKING AUTOCORELLATION WITH DIFFERENTIATION OF ORDER [1,2]

```python
import pandas as pd
from matplotlib import pyplot
from pandas.plotting import autocorrelation_plot
from pandas import Series

models_path = "../saved_models/normal/may2023"
df = pd.read_pickle("../../data/20230319_RTU_Dataset_PPC-Lab/df1.pkl")  
```

```python
# create a differenced series
def difference(dataset, interval=1):
     diff = list()
     for i in range(interval, len(dataset)):
         value = dataset[i] - dataset[i - interval]
         diff.append(value)
     return Series(diff)
```

```python
df
```

# TRYING TO FIND AN AUTOCORRELATION VALUE WITH 1-DIM DIFFERENTIATION ON MEM

```python
diff1_mem = difference(list(df["MEM_USAGE"]))
diff1_mem.plot()
```

```python
autocorrelation_plot(diff1_mem).set_xlim([20, 50])
pyplot.show()
```

## DIFFERENTIATION DIM=2 ON MEM

```python
diff2_mem = difference(diff1_mem)
diff2_mem.plot()
```

```python
autocorrelation_plot(diff2_mem).set_xlim([20, 50])
pyplot.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.layers import Bidirectional, LSTM, Input, RepeatVector, TimeDistributed
from keras.models import Model
import os
```

```python
df = pd.read_pickle("./df_reshaped.pkl")  
df.head()
```

```python
df.shape
```

```python
df_cleaned = df[["CPUs", "MEMs", "TEMPs"]]
```

```python
df_cleaned.shape
```

```python
training_size = int(df_cleaned.shape[0] * .75) 
testing_size = df_cleaned.shape[0] - training_size
print("train_size", training_size)
print("testing_size", testing_size)
train_df,test_df = df_cleaned[:training_size], df_cleaned[training_size:] 
print("train_df.shape", train_df.shape)
print("test_df.shape", test_df.shape)
```

```python
train = train_df
scalers={}
for i in train_df.columns:
    scaler = MinMaxScaler(feature_range=(-1,1))
    s_s = scaler.fit_transform(train[i].values.reshape(-1,1))
    s_s=np.reshape(s_s,len(s_s))
    scalers['scaler_'+ i] = scaler
    train[i]=s_s
test = test_df
for i in train_df.columns:
    scaler = scalers['scaler_'+i]
    s_s = scaler.transform(test[i].values.reshape(-1,1))
    s_s=np.reshape(s_s,len(s_s))
    scalers['scaler_'+i] = scaler
    test[i]=s_s
```

```python
def split_series(series, n_past, n_future):
  #
  # n_past ==> no of past observations
  #
  # n_future ==> no of future observations 
  #
  X, y = list(), list()
  for window_start in range(len(series)):
    past_end = window_start + n_past
    future_end = past_end + n_future
    if future_end > len(series):
      break
    # slicing the past and future parts of the window
    past, future = series[window_start:past_end, :], series[past_end:future_end, :]
    X.append(past)
    y.append(future)
  return np.array(X), np.array(y)
```

```python
# n_past = 10
# n_future = 5 
# n_features = 3

n_past = 35
n_future = 22 
n_features = 3
```

```python
X_train, y_train = split_series(train.values,n_past, n_future)
print("X_train.shape", X_train.shape)
print("y_train.shape", y_train.shape)
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1],n_features))
y_train = y_train.reshape((y_train.shape[0], y_train.shape[1], n_features))
print("------------ reshaping ------------")
print("X_train.shape", X_train.shape)
print("y_train.shape", y_train.shape)
print("------------ --------- ------------")
X_test, y_test = split_series(test.values,n_past, n_future)
print("X_test.shape", X_test.shape)
print("y_test.shape", y_test.shape)
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1],n_features))
y_test = y_test.reshape((y_test.shape[0], y_test.shape[1], n_features))
print("------------ reshaping ------------")
print("X_test.shape", X_test.shape)
print("y_test.shape", y_test.shape)
```

```python
# E1D1
# n_features ==> no of features at each timestep in the data.
#
encoder_inputs = Input(shape=(n_past, n_features))
encoder_l1 = LSTM(128, return_state=True)
encoder_outputs1 = encoder_l1(encoder_inputs)

encoder_states1 = encoder_outputs1[1:]

#
decoder_inputs = RepeatVector(n_future)(encoder_outputs1[0])

#
decoder_l1 = LSTM(128, return_sequences=True)(decoder_inputs,initial_state = encoder_states1)
decoder_outputs1 = TimeDistributed(tf.keras.layers.Dense(n_features))(decoder_l1)

#
model_e1d1 = Model(encoder_inputs,decoder_outputs1)

#
model_e1d1.summary()
```

```python
# E2D2
# n_features ==> no of features at each timestep in the data.
#
encoder_inputs = Input(shape=(n_past, n_features))
encoder_l1 = LSTM(128,return_sequences = True, return_state=True)
encoder_outputs1 = encoder_l1(encoder_inputs)
encoder_states1 = encoder_outputs1[1:]
encoder_l2 = LSTM(128, return_state=True)
encoder_outputs2 = encoder_l2(encoder_outputs1[0])
encoder_states2 = encoder_outputs2[1:]
#
decoder_inputs = tf.keras.layers.RepeatVector(n_future)(encoder_outputs2[0])
#
decoder_l1 = LSTM(128, return_sequences=True)(decoder_inputs,initial_state = encoder_states1)
decoder_l2 = LSTM(128, return_sequences=True)(decoder_l1,initial_state = encoder_states2)
decoder_outputs2 = TimeDistributed(tf.keras.layers.Dense(n_features))(decoder_l2)
#
model_e2d2 = Model(encoder_inputs,decoder_outputs2)
#
model_e2d2.summary()
```

```python
# E1D1_bi
# n_features ==> no of features at each timestep in the data.
#
encoder_inputs = Input(shape=(n_past, n_features))
encoder_l1 = Bidirectional(LSTM(128, return_state=True))
encoder_outputs1 = encoder_l1(encoder_inputs)

encoder_states1 = encoder_outputs1[1:]

#
decoder_inputs = RepeatVector(n_future)(encoder_outputs1[0])

#
decoder_l1 = Bidirectional(LSTM(128, return_sequences=True))(decoder_inputs,initial_state = encoder_states1)
decoder_outputs1 = TimeDistributed(tf.keras.layers.Dense(n_features))(decoder_l1)

#
model_e1d1_bi = Model(encoder_inputs,decoder_outputs1)

#
model_e1d1_bi.summary()
```

```python
reduce_lr = tf.keras.callbacks.LearningRateScheduler(lambda x: 1e-3 * 0.90 ** x)

model_e1d1.compile(optimizer=tf.keras.optimizers.Adam(), loss=tf.keras.losses.Huber())
history_e1d1=model_e1d1.fit(X_train,y_train,epochs=500,validation_data=(X_test,y_test),batch_size=32,verbose=1,callbacks=[reduce_lr])

model_e1d1_bi.compile(optimizer=tf.keras.optimizers.Adam(), loss=tf.keras.losses.Huber())
history_e1d1_bi=model_e1d1_bi.fit(X_train,y_train,epochs=500,validation_data=(X_test,y_test),batch_size=32,verbose=1,callbacks=[reduce_lr])

model_e2d2.compile(optimizer=tf.keras.optimizers.Adam(), loss=tf.keras.losses.Huber())
history_e2d2=model_e2d2.fit(X_train,y_train,epochs=500,validation_data=(X_test,y_test),batch_size=32,verbose=1,callbacks=[reduce_lr])
```

```python
pred_e1d1=model_e1d1.predict(X_test)
pred_e1d1_bi=model_e1d1_bi.predict(X_test)
pred_e2d2=model_e2d2.predict(X_test)
```

```python
for index,i in enumerate(train_df.columns):
    scaler = scalers['scaler_'+i]
    pred_e1d1[:,:,index]=scaler.inverse_transform(pred_e1d1[:,:,index])
    pred_e1d1_bi[:,:,index]=scaler.inverse_transform(pred_e1d1_bi[:,:,index])
    pred_e2d2[:,:,index]=scaler.inverse_transform(pred_e2d2[:,:,index])
    y_train[:,:,index]=scaler.inverse_transform(y_train[:,:,index])
    y_test[:,:,index]=scaler.inverse_transform(y_test[:,:,index])
```

```python
print("pred_e1d1.shape", pred_e1d1.shape)
print("pred_e1d1_bi.shape", pred_e1d1_bi.shape)
print("pred_e2d2.shape", pred_e2d2.shape)
print("y_test.shape", y_test.shape)
```
